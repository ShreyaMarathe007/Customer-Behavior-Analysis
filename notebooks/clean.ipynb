```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.columns)
```

```python
df.head()
```

```python
df.drop('unnamed:_0', axis=1, inplace=True)
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.head()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df.drop('promo_code_used', axis=1 , inplace= True)
```

```python
bins = [0, 12 , 17, 59 , df['age'].max()]
labels = ['Kid' , 'Teenager', 'Adult' , 'Senior']
df['age_group'] = pd.cut(df['age'], bins = bins, labels = labels , right = True)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Every 3 Months' : 90
}

df['purchase_cycle_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_cycle_days']].head(10)
```

```python
df['loyalty_score'] = (df['previous_purchases'] * 0.5) + \
                      (df['frequency_of_purchases'].map({'Weekly' : 5,'Fortnightly' : 4,'Bi-Weekly' : 4,'Monthly' : 3,'Quarterly' : 2,'Every 3 Months' : 2,'Annually' : 1})) + \
                      (df['subscription_status'].map({'Yes': 2, 'No': 0}))
```

```python
df['loyalty_score'].head()
```

```python
df['payment_method'].unique()
```

```python
df['payment_category'] = df['payment_method'].map({
    'Venmo': 'Digital Wallet',
    'PayPal': 'Digital Wallet',
    'Bank Transfer': 'Bank',
    'Credit Card': 'Card',
    'Debit Card': 'Card',
    'Cash': 'Cash'
})
```

```python
df['is_digital_payment'] = df['payment_method'].isin(['Venmo', 'PayPal', 'Credit Card', 'Debit Card', 'Bank Transfer'])
```

```python
df[['payment_method' , 'payment_category' , 'is_digital_payment']].head()
```
